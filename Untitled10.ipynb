```python
class Player:
  def __init__(self,claim=0):
    self.claim=claim
    self.got=0
  #getter method
  def get_claim(self):
    return self.claim
  
  #setter method
  def set_claim(self,x):
    self.claim=x

  #getter method
  def get_got(self):
    return self.got
  
  #setter method
  def set_got(self,x):
    self.got=x
  
  def addToGot(self,gotMore):
    self.got+=gotMore



def compare(p1,p2):
  if p1.get_claim()<p2.get_claim():
    return 1
  if p1.get_claim()== p2.get_claim(): 
    return 0
  else: 
    return -1

def fillTop(players, totalClaim, money):
  #check if the money is greater or equal to the claim
  if money>=totalClaim:
    for p in players:
      p.addToGot(p.get_claim()/2)
    return

  Z=money-totalClaim/2 #the money still left to divide
  N=len(players)
  #we go from right to left, that is from the most claiming person to the less claiming person or until the money has run out
  i = N-1
  while i > 0 and Z>0:
    print(i)
    a=players[i].get_claim()/2
    if i >0:
      a-=players.get[i-1].get_claim()/2
    if (a*(N-i)<=Z):
      for j in range (i,N):
        players[j].addToGot(a)
      Z-=a*(N-i)
    else:
      for j in range (i,N):
        players[j].addToGot(Z/(N-i))
      Z=0

    i = i - 1
 


def fillBottom(players, totalClaim, money):
  #we need more thenjust  the bottom vessels for dividing the money
  if(money>=totalClaim/2): 
    for p in players: 
      p.setGot(p.get_claim()/2)
    return



    

  Z=money #money left
  N=len(players);
  
  for i in range(N):
    #	check how much more can be put in this vessel
    amountLeft=players[i].get_claim()/2;
    if i>0:
      amountLeft-=players[i-1].get_claim()/2
    #check if we can top this vessel up
    if(amountLeft*(N-i))<Z:
      #top it up
      for j in range(i,N):
        players[j].addToGot(amountLeft);
      Z-=(amountLeft*(N-i))
    else: 
      # divide the remaining money equally over the remaining vessels
      for j in range(i,N):
        players[j].addToGot(Z/(N-i))
      Z=0
  



def calculateTalmud(players, money):
  #sum the total amount of claims
  totalClaim=0
  for p in players: 
    totalClaim+=p.get_claim()
  #fill the bottom half of the vessels
  fillBottom(players, totalClaim, money);

  #if we got more money then the bottom half of the vessels can indicate, also fill the top ones
  if money>totalClaim/2: 
    fillTop(players, totalClaim,money)
p0=Player(100)
p1= Player(150)
p2= Player(200)
p3= Player(210)
p4= Player(400)
p5= Player(500)
p6=Player(510)
p7=Player(520)

money=1000.0
players=[p0,p1,p2,p3,p4,p5,p6,p7]
calculateTalmud(players,money)
for p in players:
  print(p.get_got())

```
